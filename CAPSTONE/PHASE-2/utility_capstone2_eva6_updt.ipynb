# Add the Annotations to Main JSON file

```python
import json
import os
import pandas as pd
import numpy as np

for dir in os.listdir('C:\\Users\\nihar\\Downloads\\construction_materials_dataset'):
    print('currently Processing', dir)
    os.chdir('C:\\Users\\nihar\\Downloads\\construction_materials_dataset\\'+dir)
    with open('coco.json', 'r') as openfile:
    # Reading from json file
        json_object = json.load(openfile)
        print(' The original length of annotation is',len(json_object['annotations']))
    image_df = pd.DataFrame(columns=['id','image_id'])
    for dict in json_object['images']:
        image_df = image_df.append({'id':dict['id'],'image_id':dict['file_name']},ignore_index=True)
    
    df=pd.read_csv('C:\\Users\\nihar\\Documents\\theschoolofai\\EVA6\\capstone\\csv\\'+ dir + '.csv')
    seg = []
    seq = len(json_object['annotations'])
    for i in range(len(df)):
        #print('value of i is', i)
        bb = []
        seg=[]
        for value in df['bbox'][i].split(',') :
            bb.append(float(value))
        for val in df['segmentation'][i].split(','):
            seg.append(float(val))
        #print(image_df['image_id'])
        #print(df['file_name'][i] )
        iid = int(image_df[image_df['image_id'] == df['file_name'][i]]['id'])
        #print(iid, type(iid))

        json_object['annotations'].append({'id':seq,'image_id':int(iid),'category_id':int(df['cat'][i]),'segmentation':seg,
                                          'area':float(df['area'][i]),'bbox':bb})
        seq+=1
    print(' The revised length of annotation is',len(json_object['annotations']))
    out_file_name = dir + '.json'
    out_file = open(out_file_name, "w")
    json.dump(json_object, out_file, indent = 6)
    out_file.close()
    
```

# This is the easiest approach to ensure we don't get exponentially large dataset 
Find out the overlap and if it is more than 40% simply ignore the predicted one and keep the manually annotated image 

```python
def bb_intersection_over_union(boxA, boxB):
    # determine the (x, y)-coordinates of the intersection rectangle
    
    if type(boxB) == str:
        boxB = list(boxB.split(','))
#     print(type(orig), type(inf[0]))
    x0,y0,w0,h0 = boxA[0],boxA[1],boxA[2],boxA[3]
    x1,y1,w1,h1 = float(boxB[0]),float(boxB[1]),float(boxB[2]),float(boxB[3])
    
    
    xA = max(x0, x1)
    yA = max(y0, y1)
    xB = min(x0+w0, x1+w1)
    yB = min(y0+h0, y1+h1)

    # compute the area of intersection rectangle
    interArea = abs(max((xB - xA, 0)) * max((yB - yA), 0))
    if interArea == 0:
        return 0
    # compute the area of both the prediction and ground-truth
    # rectangles
    boxAArea = abs(w0 * h0)
    boxBArea = abs(w1 * h1)

    # compute the intersection over union by taking the intersection
    # area and dividing it by the sum of prediction + ground-truth
    # areas - the interesection area
    iou = interArea / float(boxAArea + boxBArea - interArea)
    # Let us see how much percentage it overlaps with the thing 
    iwt = interArea/float(boxAArea)
    # return the intersection over union value
    #print(iwt)
    return iwt
```
